```python
import sqlparse
```

```python
sql = '''CREATE EXTERNAL TABLE `dwd.dw_jobs_headhunting_consultant_job_candidate_da` (
    `candidate_progress_id` BIGINT COMMENT '自增主键id',
    `uid` BIGINT COMMENT 'uid',
    `job_req_id` BIGINT COMMENT 'ai_ht_job_requirements的id',
    `candidate_frm` STRING COMMENT '候选人来源',
    `candidate_uid` BIGINT COMMENT '候选人的uid',
    `emobile` STRING COMMENT '加密后的手机号用于查重',
    `current_progress` TINYINT COMMENT '当前进度',
    `create_ts` TIMESTAMP COMMENT '创建时间',
    `update_ts` TIMESTAMP COMMENT '更新时间',
    `candidate_status` BIGINT COMMENT '候选人状态 1 已保存，2 沟通中，3 已锁定（其他职位在推荐流程中），4 已推荐（流程中），5 已终止，6 已成单',
    `is_starred` TINYINT COMMENT '是否标记为非常匹配',
    `action_time` TIMESTAMP COMMENT '操作时间',
    `ai_consultant_mm_uid` BIGINT COMMENT 'ai_consultant_mm_uid',
    `ai_consultant_wechat_uid` STRING COMMENT 'ai_consultant的微信uid',
    `candidate_wechat_uid` STRING COMMENT 'candidate_uid的微信uid',
    `phone_call_ts` TIMESTAMP COMMENT '拨打虚拟电话时间',
    `wechat_status` TINYINT COMMENT '微信消息状态',
    `latest_wechat_ts` TIMESTAMP COMMENT '最近一次微信消息时间',
    `mmim_status` TINYINT COMMENT '脉脉IM消息状态',
    `latest_mmim_ts` TIMESTAMP COMMENT '最近一次脉脉IM消息时间',
    `ai_action_status` TINYINT COMMENT 'AI/策略状态',
    `latest_ai_action_ts` TIMESTAMP COMMENT '最近一次AI行为时间',
    `contact_status` TINYINT COMMENT '沟通状态：0 初始状态，1 破冰，2 脉脉IM聊天，3 微信加好友，4 微信聊天，5 待电话沟通，11 连接失败，12 无意向，13 重复推荐'
) COMMENT 'AI猎头-候选人表'
PARTITIONED BY ( `p_date` string COMMENT '日期')
ROW FORMAT DELIMITED FIELDS TERMINATED BY '\t'
STORED AS ORC
TBLPROPERTIES ('orc.compression'='snappy');'''
```
